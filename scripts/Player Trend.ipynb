```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/Ryan/Documents/air_yards_model/data/airyards_data.csv')
df
```

```python
df_kupp = df[(df['name'] == 'Cooper Kupp') & (df['season'] == 2021)]
plt.plot(df_kupp['week'], df_kupp['actual_fantasy_pts'], label = 'actual fantasy pts')
plt.plot(df_kupp['week'], df_kupp['season_avg_fantasy_points_ppr'], label = 'season avg fantasy pts')
plt.legend()
plt.show()
```
